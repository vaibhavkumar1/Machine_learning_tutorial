```python
x=df_google_playstore_app['Rating']
y=df_google_playstore_app['Reviews']

plt.title('Google playstore apps scatter plot')
plt.xlabel("Ratings",fontsize=15)
plt.ylabel('Reviews',fontsize=15)
plt.scatter(x,y,label='Review and Rating')
plt.legend(loc=2)
```

```python
x=df_google_playstore_app['Rating']
y=df_google_playstore_app['Reviews']

plt.figure(figsize=(15,10))
plt.title('Google playstore apps scatter plot')
plt.xlabel("Ratings",fontsize=15)
plt.ylabel('Reviews',fontsize=15)
plt.scatter(x,y,label='Review and Rating',c='r',marker="*",alpha=0.5,linewidth=10)
plt.legend(loc=2)
```
