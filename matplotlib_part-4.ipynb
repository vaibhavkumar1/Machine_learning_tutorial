```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
import random
```

```python
ml_student_age=np.random.randint(18,45,(100))
py_student_age=np.random.randint(15,440,(100))


print(ml_student_age)
print(py_student_age)
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_student_age,bins,rwidth=0.9,histtype="step")
plt.title("ML Student age histogram")
plt.xlabel("Age category")
plt.ylabel("No. of student")
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]

plt.figure(figsize=(10,9))
plt.hist([ml_student_age,py_student_age],bins,rwidth=0.9,histtype="bar",orientation="vertical",color=["m","y"],label=["ML Student",
                                                                                                                      "Python Student"])

#plt.hist(py_student_age,bins,rwidth=0.9,histtype="bar",orientation="vertical",color="y",label="Python Student")


plt.legend(loc=2)
plt.title("ML Student age histogram")
plt.xlabel("Age category")
plt.ylabel("No. of student")
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.style.use("ggplot")

plt.figure(figsize=(10,10))
plt.hist([ml_student_age,py_student_age],bins,rwidth=0.9,histtype="bar",orientation="vertical",color=["m","y"],label=["ML Student",
                                                                                                                      "Python Student"])

#plt.hist(py_student_age,bins,rwidth=0.9,histtype="bar",orientation="vertical",color="y",label="Python Student")


plt.legend(loc=2)
plt.title("ML Student age histogram")
plt.xlabel("Age category")
plt.ylabel("No. of student")
plt.show()
```
