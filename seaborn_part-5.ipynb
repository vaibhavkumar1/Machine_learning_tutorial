```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
```

```python
sns.distplot(df['total_bill'])
```

```python
sns.set()
sns.distplot(df['total_bill'])
```

```python
plt.figure(figsize=(10,8))
sns.set()
sns.distplot(df['total_bill'])
```

```python
bins=[1,5,10,15,20,25,30,35,40,45,50,55,60]

plt.figure(figsize=(10,8))
sns.set()
sns.distplot(df['total_bill'],bins=bins)
plt.show()
```

```python
bins=[1,5,10,15,20,25,30,35,40,45,50,55,60]

plt.figure(figsize=(10,8))
sns.set()
sns.distplot(df['total_bill'],bins=bins,hist_kws={"color":"pink","edgecolor":"yellow"})
plt.show()
```

```python
bins=[1,5,10,15,20,25,30,35,40,45,50,55,60]

plt.figure(figsize=(10,8))
sns.set()
sns.distplot(df['total_bill'],bins=bins,hist_kws={"linewidth":"3","color":"pink","edgecolor":"yellow"})
plt.show()
```
