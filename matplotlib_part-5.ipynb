# ploting bar chart

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
from matplotlib import style
```

```python
classes=["Python","ML","AI","DS","R"]
class1=[30,10,20,25,10]
class2=[40,5,20,20,10]
class3=[35,5,30,15,15]
```

```python
plt.bar(classes,class1)
```

```python
plt.barh(classes,class1)
```

```python
plt.bar(classes,class1,width=0.2,align="edge",color="y",edgecolor="m",linewidth=5,alpha=0.9,linestyle="--",
        label="Class1 student")
```

```python
style.use("ggplot")
plt.figure(figsize=(10,8))
plt.bar(classes,class1,width=0.5,align="edge",color="black",edgecolor="m",linewidth=3,alpha=0.9,linestyle="--",
        label="Class1 student")
```

```python
style.use("ggplot")
plt.figure(figsize=(10,8))
plt.bar(classes,class1,width=0.5,align="edge",color="black",edgecolor="m",linewidth=3,alpha=0.9,linestyle="--",
        label="Class1 student")

plt.title("Bar chart of IAIP class",fontsize=15)
plt.xlabel("Number of student",fontsize=15)
plt.ylabel("Classes",fontsize=15)
```

```python
style.use("ggplot")
plt.figure(figsize=(10,8))
plt.bar(classes,class1,width=0.2,align="edge",color="black",
        label="Class1 student")
plt.bar(classes,class2,width=0.2,align="edge",color="green",
        label="Class2 student")

plt.title("Bar chart of IAIP class",fontsize=15)
plt.xlabel("Number of student",fontsize=15)
plt.ylabel("Classes",fontsize=15)
```

```python
style.use("ggplot")

classes_index=np.arange(len(classes))
plt.figure(figsize=(10,8))
width=0.2
plt.bar(classes_index,class1,width,align="edge",color="black",
        label="Class1 student")
plt.bar(classes_index+width,class2,width,align="edge",color="green",
        label="Class2 student")
plt.bar(classes_index+width+width,class3,width,align="edge",color="y",
        label="Class3 student")


plt.xticks(classes_index+width,classes,rotation=20)

plt.title("Bar chart of IAIP class",fontsize=15)
plt.xlabel("Number of student",fontsize=15)
plt.ylabel("Classes",fontsize=15)
```

```python
style.use("ggplot")

classes_index=np.arange(len(classes))
plt.figure(figsize=(10,8))
width=0.2
plt.barh(classes_index,class1,width,align="edge",color="black",
        label="Class1 student")
plt.barh(classes_index+width,class2,width,align="edge",color="green",
        label="Class2 student")
plt.barh(classes_index+width+width,class3,width,align="edge",color="y",
        label="Class3 student")


plt.yticks(classes_index+width,classes,rotation=20)

plt.legend(loc=1)
plt.title("Bar chart of IAIP class",fontsize=15)
plt.ylabel("Number of student",fontsize=15)
plt.xlabel("Classes",fontsize=15)
```
