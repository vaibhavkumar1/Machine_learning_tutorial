```python
import numpy as np
import matplotlib.pyplot as v
a=np.arange(0,10,.1)
b1=[np.sin(i)for i in a]
b2=[np.cos(i)for i in a]

v.figure(1)

v.plot(a,b1,'r')
v.title("graph of sin")
v.xlabel("value opf a")
v.ylabel("sin")
for i in range(0,10):
    v.text(i,0.50,"_")

v.figure(2)

v.plot(a,b2,"b")
v.title("graph of cos")
v.xlabel("value opf a")
v.ylabel("cos")
for i in range(0,10):
    v.text(i,.75,"_")

v.show()
```
