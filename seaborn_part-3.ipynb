```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
tips_df=sns.load_dataset("tips")
tips_df
```

```python
sns.lineplot(x='size',y='total_bill', data=tips_df,hue="sex",style="sex",palette="Accent")
plt.show()
```

```python
sns.lineplot(x='size',y='total_bill', data=tips_df,hue="sex",style="sex",palette="hot",dashes=False,markers=['o','<'],legend=False)
plt.show()
```

```python
plt.figure(figsize=(10,9))
sns.set(style="darkgrid")
sns.lineplot(x='size',y='total_bill', data=tips_df,hue="sex",style="sex"
             ,palette="hot",dashes=False,markers=['o','<'],legend="brief")
plt.xlabel("Size" ,fontsize=20)
plt.ylabel("Total_bill" ,fontsize=20)
plt.savefig("hello.png")

plt.show()
```
