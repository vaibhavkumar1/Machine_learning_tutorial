```python
import matplotlib.pyplot as plt
```

```python
classes=['python','React','javascript','Data science','Machine learning']
class1=[45,50,15,30,35]
```

```python
plt.pie([1,2,3,10])
plt.show()
```

```python
plt.figure(figsize=(10,9))
plt.pie(class1,labels=classes)
plt.show()
```

```python
plt.figure(figsize=(6,9))
textprops={"fontsize":15}
plt.pie(class1,labels=classes,explode=[0,0,0.2,0,0.1],colors=['green','r','b','y','c'],autopct='%0.2f%%',shadow=True
       ,radius=1.4,startangle=0,textprops=textprops)
plt.show()
```

```python
plt.figure(figsize=(6,9))
textprops={"fontsize":15}
plt.pie(class1,labels=classes,explode=[0,0,0.2,0,0.1],colors=['green','r','b','y','c'],autopct='%0.2f%%',shadow=True
       ,radius=1.4,startangle=0,textprops=textprops)
plt.legend(loc=1)
plt.show()
```
